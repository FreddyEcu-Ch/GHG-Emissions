```python
new_s = {"E5 (avg) EC CO2-A (71 t/ha)": "E5 (avg)\nEC CO2-A \n (71 t/ha)", "E5 (avg) EC CO2-A (86 t/ha)": "E5 (avg)\nEC CO2-A\n(86 t/ha)", 
        "E5 (no disp.) EC CO2-A": "E5 (no disp.)\nEC CO2-A", "E5 (mgl) EC CO2-A": "E5 (mgl)\nEC CO2-A"}

df_2['Scenarios'] = df_2['Scenarios'].replace(new_s)
```

```python
def bar_plots(df, row_i, row_f, text):
    fig, ax = plt.subplots(figsize=(24, 14))
    df.iloc[row_i:row_f, :-1].plot(x='Scenarios', kind='bar', stacked=True, ax=ax)
    ax.set_xticklabels(ax.get_xticklabels())
    plt.rcParams['font.size'] = 24
    ax.set_xlabel(None)
    ax.set_ylabel(text, fontsize=24)
    plt.yticks(fontsize=24)
    ax.set_ylim(df.iloc[row_i: row_f, 1:].min().min())
    ax.spines['top'].set_position(('data', 0))
    ax.legend(loc="upper center", ncol=len(df.columns), fontsize=17)
    fig.savefig("fig.png", pad_inches=0)
    plt.show()
```
